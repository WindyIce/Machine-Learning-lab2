```python
import numpy as np
from sklearn.datasets import load_svmlight_file
import scipy.sparse
import matplotlib.pyplot as plt

# load the data file
X_train, Y_train = load_svmlight_file('a9a.txt')
X_test, Y_test = load_svmlight_file('a9a.t')

# split the data into training set and validation set
X_train = np.asarray(scipy.sparse.csr_matrix(X_train).todense())
Y_train = np.asarray(scipy.sparse.csr_matrix(Y_train).todense())
X_test = np.asarray(scipy.sparse.csr_matrix(X_test).todense())
Y_test = np.asarray(scipy.sparse.csr_matrix(Y_test).todense())

row = X_train.shape[0]
column = X_train.shape[1]
testrow = X_test.shape[0]
testcolumn = X_test.shape[1]

# Initialize the parameter
X = np.hstack((X_train, np.ones((row, 1))))  # Let the last column in X to be 1
x = np.hstack((X_test, np.ones((testrow, 1))))
x = np.hstack((x, np.ones((testrow, 1))))
learning_rate = 0.0001
patchNum=100
# NAG parameters
W_NAG = np.zeros((column + 1, 1))  # Merge the W and b
V_NAG = np.zeros((column + 1, 1))
momentum_NAG = 0.5
Vt = np.zeros((column + 1, 1))
# RMSProp parameters
W_RMSProp = np.zeros((column + 1, 1))
steadyNum_RMSProp = 0.001
decreaseSpeed = 0.2
R_RMSProp = 0
# AdaDelta parameters
W_AdaDelta = np.zeros((column + 1, 1))
steadyNum_AdaDelta = 0.0001
R_AdaDelta = 0
# Adam parameters
W_Adam = np.zeros((column + 1, 1))
steadyNum_Adam = 0.0001
momentum1_Adam = 0.9
momentum2_Adam = 0.999
S_Adam = np.zeros((column + 1, 1))
R_Adam = 0


gradient_rounds = 10000  # rounds for training
xplot = []
yplot_NAG = []
yplotV_NAG = []
yplotLoss_NAG = []
yplotVLoss_NAG = []
yplot_RMSProp = []
yplotLoss_RMSProp = []
yplotV_RMSProp = []
yplotVLoss_RMSProp = []
yplot_AdaDelta = []
yplotLoss_AdaDelta = []
yplotV_AdaDelta = []
yplotVLoss_AdaDelta = []
yplot_Adam = []
yplotLoss_Adam = []
yplotV_Adam = []
yplotVLoss_Adam = []

for t in range(gradient_rounds):
    # gradient descent
    random_num = np.random.random_integers(row - 1)  # The random number for Stochastic gradient descent
    it = np.reshape(X[random_num], (1, column + 1))  # The random row
    gradient=np.reshape(np.dot(Y_train.T[random_num], it), (column + 1, 1))
    if (Y_train.T[random_num] * (np.dot(W_NAG.T, it.T))[0]) < 1:  # max(0, Y_train.T[random_num] * (np.dot(W.T, it.T)))
        # NAG
        V_NAG = momentum_NAG * V_NAG + learning_rate * gradient
        W_NAG = W_NAG + V_NAG
    if (Y_train.T[random_num] * (np.dot(W_RMSProp.T, it.T))[0]) < 1:  # max(0, Y_train.T[random_num] * (np.dot(W.T, it.T)))
        # RMSProp
        R_RMSProp = decreaseSpeed * R_RMSProp + (1 - decreaseSpeed) * np.dot(gradient.T, gradient)
        change_RMSProp = learning_rate * (row / 2) / (steadyNum_RMSProp + np.sqrt(R_RMSProp)) * gradient
        W_RMSProp = W_RMSProp + change_RMSProp
    if (Y_train.T[random_num] * (np.dot(W_AdaDelta.T, it.T))[0]) < 1:
        # AdaDelta
        R_AdaDelta = R_AdaDelta + np.dot(gradient.T, gradient)
        change_AdaDelta = learning_rate * (row / 2) / (steadyNum_AdaDelta + np.sqrt(R_AdaDelta)) * gradient
        W_AdaDelta = W_AdaDelta + change_AdaDelta
    if (Y_train.T[random_num] * (np.dot(W_Adam.T, it.T))[0]) < 1:
        # Adam
        S_Adam = momentum1_Adam * S_Adam + (1 - momentum1_Adam) * gradient
        R_Adam = momentum2_Adam * R_Adam + (1 - momentum2_Adam) * np.dot(gradient.T, gradient)
        S_predict = S_Adam / (1 - momentum1_Adam)
        R_predict = R_Adam / (1 - momentum2_Adam)
        change_Adam = (learning_rate / (np.sqrt(R_predict) + steadyNum_Adam) * S_predict)
        W_Adam = W_Adam + change_Adam
    if(t%100==0):
        # Computing the correct
        xplot.append(t)
        correct_NAG = 0
        loss_NAG = 0
        correct_RMSProp = 0
        loss_RMSProp = 0
        correct_AdaDelta = 0
        loss_AdaDelta = 0
        correct_Adam = 0
        loss_Adam = 0
        # Compute the correct rate in train set
        for i in range(row):
            loss_NAG += max(0, 1 - Y_train.T[i] * np.dot(W_NAG.T, X[i]))
            loss_RMSProp += max(0, 1 - Y_train.T[i] * np.dot(W_RMSProp.T, X[i]))
            loss_AdaDelta += max(0, 1 - Y_train.T[i] * np.dot(W_AdaDelta.T, X[i]))
            loss_Adam += max(0, 1 - Y_train.T[i] * np.dot(W_Adam.T, X[i]))
            if ((np.dot(X[i], W_NAG)[0] > 0) == (Y_train.T[i] > 0)):
                correct_NAG += 1
            if ((np.dot(X[i], W_RMSProp)[0] > 0) == (Y_train.T[i] > 0)):
                correct_RMSProp += 1
            if ((np.dot(X[i], W_AdaDelta)[0] > 0) == (Y_train.T[i] > 0)):
                correct_AdaDelta += 1
            if ((np.dot(X[i], W_Adam)[0] > 0) == (Y_train.T[i] > 0)):
                correct_Adam += 1
        yplot_NAG.append(correct_NAG / row)
        yplotLoss_NAG.append(loss_NAG / row)
        yplot_RMSProp.append(correct_RMSProp / row)
        yplotLoss_RMSProp.append(loss_RMSProp / row)
        yplot_AdaDelta.append(correct_AdaDelta / row)
        yplotLoss_AdaDelta.append(loss_AdaDelta / row)
        yplot_Adam.append(correct_Adam / row)
        yplotLoss_Adam.append(loss_Adam / row)
        correct_NAG = 0
        loss_NAG = 0
        correct_RMSProp = 0
        loss_RMSProp = 0
        correct_AdaDelta = 0
        loss_AdaDelta = 0
        correct_Adam = 0
        loss_Adam = 0
        # Compute the correct rate in validation set
        for i in range(testrow):
            eachx = X_test[i]
            x1 = np.hstack((eachx, 1))
            x1 = np.hstack((x1, 1))
            loss_NAG += max(0, 1 - Y_test.T[i] * np.dot(W_NAG.T, x1))
            loss_RMSProp += max(0, 1 - Y_test.T[i] * np.dot(W_RMSProp.T, x1))
            loss_AdaDelta += max(0, 1 - Y_test.T[i] * np.dot(W_AdaDelta.T, x1))
            loss_Adam += max(0, 1 - Y_test.T[i] * np.dot(W_Adam.T, x1))
            if ((np.dot(x1, W_NAG)[0] > 0) == (Y_test.T[i] > 0)):
                correct_NAG += 1  # Hit!
            if ((np.dot(x1, W_RMSProp)[0] > 0) == (Y_test.T[i] > 0)):
                correct_RMSProp += 1  # Hit!
            if ((np.dot(x1, W_AdaDelta)[0] > 0) == (Y_test.T[i] > 0)):
                correct_AdaDelta += 1  # Hit!
            if ((np.dot(x1, W_Adam)[0] > 0) == (Y_test.T[i] > 0)):
                correct_Adam += 1  # Hit!
        yplotV_NAG.append(correct_NAG / testrow)
        yplotVLoss_NAG.append(loss_NAG / testrow)
        yplotV_RMSProp.append(correct_RMSProp / testrow)
        yplotVLoss_RMSProp.append(loss_RMSProp / testrow)
        yplotV_AdaDelta.append(correct_AdaDelta / testrow)
        yplotVLoss_AdaDelta.append(loss_AdaDelta / testrow)
        yplotV_Adam.append(correct_Adam / testrow)
        yplotVLoss_Adam.append(loss_Adam / testrow)



plt.title('During the training')
plt.xlabel('Rounds')
plt.ylabel('Hit Rate')
plt.plot(np.array(xplot), np.array(yplot_NAG), color="blue", linewidth=1.0, linestyle="-", label="Train Set NAG")
plt.plot(np.array(xplot), np.array(yplotV_NAG), color="red", linewidth=1.0, linestyle="-", label="Validation Set NAG")
plt.plot(np.array(xplot), np.array(yplot_RMSProp), color="green", linewidth=1.0, linestyle="-",
         label="Train Set RMSProp")
plt.plot(np.array(xplot), np.array(yplotV_RMSProp), color="purple", linewidth=1.0, linestyle="-",
         label="Validation Set RMSProp")
plt.plot(np.array(xplot), np.array(yplot_AdaDelta), color="black", linewidth=1.0, linestyle="-",
         label="Train Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplotV_AdaDelta), color="grey", linewidth=1.0, linestyle="-",
         label="Validation Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplot_Adam), color="yellow", linewidth=1.0, linestyle="-", label="Train Set Adam")
plt.plot(np.array(xplot), np.array(yplotV_Adam), color="pink", linewidth=1.0, linestyle="-", label="Validation Set Adam")
plt.legend(loc='best')
plt.figure(2)
plt.subplot(111)
plt.title('During the training')
plt.xlabel('Rounds')
plt.ylabel('Loss Value')
plt.plot(np.array(xplot), np.array(yplotLoss_NAG), color="blue", linewidth=1.0, linestyle="-", label="Train Set NAG")
plt.plot(np.array(xplot), np.array(yplotVLoss_NAG), color="red", linewidth=1.0, linestyle="-",
         label="Validation Set NAG")
plt.plot(np.array(xplot), np.array(yplotLoss_RMSProp), color="green", linewidth=1.0, linestyle="-",
         label="Train Set RMSProp")
plt.plot(np.array(xplot), np.array(yplotVLoss_RMSProp), color="purple", linewidth=1.0, linestyle="-",
         label="Validation Set RMSProp")
plt.plot(np.array(xplot), np.array(yplotLoss_AdaDelta), color="black", linewidth=1.0, linestyle="-",
         label="Train Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplotVLoss_AdaDelta), color="grey", linewidth=1.0, linestyle="-",
         label="Validation Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplotLoss_Adam), color="yellow", linewidth=1.0, linestyle="-", label="Train Set Adam")
plt.plot(np.array(xplot), np.array(yplotVLoss_Adam), color="pink", linewidth=1.0, linestyle="-",
         label="Validation Set Adam")
plt.legend(loc='best')
plt.show()
```
