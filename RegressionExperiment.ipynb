```python
import numpy as np
from sklearn.datasets import load_svmlight_file
import scipy.sparse
import matplotlib.pyplot as plt

# load the data file
X_train, Y_train = load_svmlight_file('a9a.txt')
X_test, Y_test = load_svmlight_file('a9a.t')

# split the data into training set and validation set
X_train = np.asarray(scipy.sparse.csr_matrix(X_train).todense())
Y_train = np.asarray(scipy.sparse.csr_matrix(Y_train).todense())
X_test = np.asarray(scipy.sparse.csr_matrix(X_test).todense())
Y_test = np.asarray(scipy.sparse.csr_matrix(Y_test).todense())

row = X_train.shape[0]
column = X_train.shape[1]
testrow = X_test.shape[0]
testcolumn = X_test.shape[1]

# Initialize the parameter
X = np.hstack((X_train, np.ones((row, 1))))  # Let the last column in X to be 1
x = np.hstack((X_test, np.ones((testrow, 1))))
x = np.hstack((x, np.ones((testrow, 1))))
learning_rate = 0.0001
# NAG parameters
W_NAG = np.zeros((column + 1, 1))  # Merge the W and b
V_NAG = np.zeros((column + 1, 1))
momentum_NAG = 0.5
Vt = np.zeros((column + 1, 1))
# RMSProp parameters
W_RMSProp = np.zeros((column + 1, 1))
steadyNum_RMSProp = 0.0001
decreaseSpeed = 0.9
R_RMSProp = 0
# AdaDelta parameters
W_AdaDelta = np.zeros((column + 1, 1))
steadyNum_AdaDelta = 0.0001
R_AdaDelta = 0
# Adam parameters
W_Adam = np.zeros((column + 1, 1))
steadyNum_Adam = 0.0001
momentum1_Adam = 0.9
momentum2_Adam = 0.999
S_Adam = np.zeros((column + 1, 1))
R_Adam = 0

gradient_rounds = 100  # rounds for training
threshold = 0

# pyplot initial
plt.figure(1)
plt.subplot(111)
xplot = []
yplotNAG = []
yplotLossNAG = []
yplotVNAG = []
yplotVLossNAG = []
yplotRMSProp = []
yplotLossRMSProp = []
yplotVRMSProp = []
yplotVLossRMSProp = []
yplotAdaDelta = []
yplotLossAdaDelta = []
yplotVAdaDelta = []
yplotVLossAdaDelta = []
yplotAdam = []
yplotLossAdam = []
yplotVAdam = []
yplotVLossAdam = []

# define the sigmond function
def sigmond(X):
    return 2.0 / (1 + np.exp(-X)) - 1


for i in range(gradient_rounds):

    predict = sigmond(np.dot(X, W_NAG))
    wrong = Y_train.T - predict
    gradient = np.dot(X.T, wrong)
    # NAG
    V_NAG = momentum_NAG * V_NAG + learning_rate * gradient
    W_NAG = W_NAG + V_NAG
    # RMSProp
    R_RMSProp = decreaseSpeed * R_RMSProp + (1 - decreaseSpeed) * np.dot(gradient.T, gradient)
    change_RMSProp = learning_rate * (row / 2) / (steadyNum_RMSProp + np.sqrt(R_RMSProp)) * gradient
    W_RMSProp = W_RMSProp + change_RMSProp
    # AdaDelta
    R_AdaDelta = R_AdaDelta + np.dot(gradient.T, gradient)
    change_AdaDelta = learning_rate * (row / 2) / (steadyNum_AdaDelta + np.sqrt(R_AdaDelta)) * gradient
    W_AdaDelta = W_AdaDelta + change_AdaDelta
    # Adam
    S_Adam = momentum1_Adam * S_Adam + (1 - momentum1_Adam) * gradient
    R_Adam = momentum2_Adam * R_Adam + (1 - momentum2_Adam) * np.dot(gradient.T, gradient)
    S_predict = S_Adam / (1 - momentum1_Adam)
    R_predict = R_Adam / (1 - momentum2_Adam)
    change_Adam = (learning_rate / (np.sqrt(R_predict) + steadyNum_Adam) * S_predict)
    W_Adam = W_Adam + change_Adam
    if (i % 2 == 0):
        xplot.append(i)
        correct_NAG = 0
        loss_NAG = 0
        correct_RMSProp = 0
        loss_RMSProp = 0
        correct_AdaDelta = 0
        loss_AdaDelta = 0
        correct_Adam = 0
        loss_Adam = 0
        for j in range(row):
            # Compute the loss in training set
            loss_NAG += np.log(1 + np.exp(-(Y_train.T[j] * np.dot(W_NAG.T, X[j]))))
            loss_RMSProp += np.log(1 + np.exp(-(Y_train.T[j] * np.dot(W_RMSProp.T, X[j]))))
            loss_AdaDelta += np.log(1 + np.exp(-(Y_train.T[j] * np.dot(W_AdaDelta.T, X[j]))))
            loss_Adam += np.log(1 + np.exp(-(Y_train.T[j] * np.dot(W_AdaDelta.T, X[j]))))/np.e
            # Compute the hit rate in training set
            if (sigmond(np.reshape(np.dot(X[j], W_NAG), (1, 1)))[0, 0] < threshold) and (Y_train.T[j] == -1):
                correct_NAG += 1
            if (sigmond(np.reshape(np.dot(X[j], W_NAG), (1, 1)))[0, 0] >= threshold) and (Y_train.T[j] == 1):
                correct_NAG += 1
            if (sigmond(np.reshape(np.dot(X[j], W_RMSProp), (1, 1)))[0, 0] < threshold) and (Y_train.T[j] == -1):
                correct_RMSProp += 1
            if (sigmond(np.reshape(np.dot(X[j], W_RMSProp), (1, 1)))[0, 0] >= threshold) and (Y_train.T[j] == 1):
                correct_RMSProp += 1
            if (sigmond(np.reshape(np.dot(X[j], W_AdaDelta), (1, 1)))[0, 0] < threshold) and (Y_train.T[j] == -1):
                correct_AdaDelta += 1
            if (sigmond(np.reshape(np.dot(X[j], W_AdaDelta), (1, 1)))[0, 0] >= threshold) and (Y_train.T[j] == 1):
                correct_AdaDelta += 1
            if (sigmond(np.reshape(np.dot(X[j], W_Adam), (1, 1)))[0, 0] < threshold) and (Y_train.T[j] == -1):
                correct_Adam += 1
            if (sigmond(np.reshape(np.dot(X[j], W_Adam), (1, 1)))[0, 0] >= threshold) and (Y_train.T[j] == 1):
                correct_Adam += 1
        yplotNAG.append(correct_NAG / row)
        yplotLossNAG.append(loss_NAG / row)
        yplotRMSProp.append(correct_RMSProp / row)
        yplotLossRMSProp.append(loss_RMSProp / row)
        yplotAdaDelta.append(correct_AdaDelta / row)
        yplotLossAdaDelta.append(loss_AdaDelta / row)
        yplotAdam.append(correct_Adam / row)
        yplotLossAdam.append(loss_Adam / row)
        correct_NAG = 0
        loss_NAG = 0
        correct_RMSProp = 0
        loss_RMSProp = 0
        correct_AdaDelta = 0
        loss_AdaDelta = 0
        correct_Adam = 0
        loss_Adam = 0
        for j in range(testrow):
            # Compute the loss in validation set
            loss_NAG += np.log(1 + np.exp(-(Y_test.T[j] * np.dot(W_NAG.T, x[j]))))
            loss_RMSProp += np.log(1 + np.exp(-(Y_test.T[j] * np.dot(W_RMSProp.T, x[j]))))
            loss_AdaDelta += np.log(1 + np.exp(-(Y_test.T[j] * np.dot(W_AdaDelta.T, x[j]))))
            loss_Adam += np.log(1 + np.exp(-(Y_test.T[j] * np.dot(W_AdaDelta.T, x[j]))))/np.e
            # Compute the hit rate in validation set
            if (sigmond(np.reshape(np.dot(x[j], W_NAG), (1, 1)))[0, 0] < threshold) and (Y_test.T[j] == -1):
                correct_NAG += 1
            if (sigmond(np.reshape(np.dot(x[j], W_NAG), (1, 1)))[0, 0] >= threshold) and (Y_test.T[j] == 1):
                correct_NAG += 1
            if (sigmond(np.reshape(np.dot(x[j], W_RMSProp), (1, 1)))[0, 0] < threshold) and (Y_test.T[j] == -1):
                correct_RMSProp += 1
            if (sigmond(np.reshape(np.dot(x[j], W_RMSProp), (1, 1)))[0, 0] >= threshold) and (Y_test.T[j] == 1):
                correct_RMSProp += 1
            if (sigmond(np.reshape(np.dot(x[j], W_AdaDelta), (1, 1)))[0, 0] < threshold) and (Y_test.T[j] == -1):
                correct_AdaDelta += 1
            if (sigmond(np.reshape(np.dot(x[j], W_AdaDelta), (1, 1)))[0, 0] >= threshold) and (Y_test.T[j] == 1):
                correct_AdaDelta += 1
            if (sigmond(np.reshape(np.dot(x[j], W_Adam), (1, 1)))[0, 0] < threshold) and (Y_test.T[j] == -1):
                correct_Adam += 1
            if (sigmond(np.reshape(np.dot(x[j], W_Adam), (1, 1)))[0, 0] >= threshold) and (Y_test.T[j] == 1):
                correct_Adam += 1
        yplotVNAG.append(correct_NAG / testrow)
        yplotVLossNAG.append(loss_NAG / testrow)
        yplotVRMSProp.append(correct_RMSProp / testrow)
        yplotVLossRMSProp.append(loss_RMSProp / testrow)
        yplotVAdaDelta.append(correct_AdaDelta / testrow)
        yplotVLossAdaDelta.append(loss_AdaDelta / testrow)
        yplotVAdam.append(correct_Adam / testrow)
        yplotVLossAdam.append(loss_Adam / testrow)
plt.title('During the training')
plt.xlabel('Rounds')
plt.ylabel('Hit Rate')
plt.plot(np.array(xplot), np.array(yplotNAG), color="blue", linewidth=1.0, linestyle="-", label="Train Set NAG")
plt.plot(np.array(xplot), np.array(yplotVNAG), color="red", linewidth=1.0, linestyle="-", label="Validation Set NAG")
plt.plot(np.array(xplot), np.array(yplotRMSProp), color="green", linewidth=1.0, linestyle="-",
         label="Train Set RMSProp")
plt.plot(np.array(xplot), np.array(yplotVRMSProp), color="purple", linewidth=1.0, linestyle="-",
         label="Validation Set RMSProp")
plt.plot(np.array(xplot), np.array(yplotAdaDelta), color="black", linewidth=1.0, linestyle="-",
         label="Train Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplotVAdaDelta), color="grey", linewidth=1.0, linestyle="-",
         label="Validation Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplotAdam), color="yellow", linewidth=1.0, linestyle="-", label="Train Set Adam")
plt.plot(np.array(xplot), np.array(yplotVAdam), color="pink", linewidth=1.0, linestyle="-", label="Validation Set Adam")
plt.legend(loc='best')
plt.figure(2)
plt.subplot(111)
plt.title('During the training')
plt.xlabel('Rounds')
plt.ylabel('Loss Value')
plt.plot(np.array(xplot), np.array(yplotLossNAG), color="blue", linewidth=1.0, linestyle="-", label="Train Set NAG")
plt.plot(np.array(xplot), np.array(yplotVLossNAG), color="red", linewidth=1.0, linestyle="-",
         label="Validation Set NAG")
plt.plot(np.array(xplot), np.array(yplotLossRMSProp), color="green", linewidth=1.0, linestyle="-",
         label="Train Set RMSProp")
plt.plot(np.array(xplot), np.array(yplotVLossRMSProp), color="purple", linewidth=1.0, linestyle="-",
         label="Validation Set RMSProp")
plt.plot(np.array(xplot), np.array(yplotLossAdaDelta), color="black", linewidth=1.0, linestyle="-",
         label="Train Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplotVLossAdaDelta), color="grey", linewidth=1.0, linestyle="-",
         label="Validation Set AdaDelta")
plt.plot(np.array(xplot), np.array(yplotLossAdam), color="yellow", linewidth=1.0, linestyle="-", label="Train Set Adam")
plt.plot(np.array(xplot), np.array(yplotVLossAdam), color="pink", linewidth=1.0, linestyle="-",
         label="Validation Set Adam")
plt.legend(loc='best')
plt.show()
```
